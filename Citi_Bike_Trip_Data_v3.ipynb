##### Create Functions to convert datetimes from one data type to another

```python
# # Function to convert datetime (dtype: object) to datetime (dtype: datetime64)
# def convertDateTimeToDateTime64(inputDateTime):
#     outputDateTime = pd.to_datetime(inputDateTime)
#     return outputDateTime

# # Function to convert datetime (dtype: datetime64) to unix/epoch
# def convertDateTimeToUnix(inputDateTime):
#     outputDateTime = (inputDateTime - dt.datetime(1970,1,1)).dt.total_seconds()
#     return outputDateTime

# Function to convert datetime (dtype: datetime64) to unix/epoch
def convertDateTimeToUnix(inputDateTime):
    outputDateTime = (pd.to_datetime(inputDateTime) - dt.datetime(1970,1,1)).dt.total_seconds()
    return outputDateTime
 
# # Another function to convert datetime (dtype: datetime64) to unix/epoch (less accurate)
# def convertDateTimeToUnix2(inputDateTime):
#     outputDateTime = inputDateTime.astype("int64")//1e9
#     return outputDateTime

# # Another function to convert datetime (dtype: datetime64) to unix/epoch (less accurate)
# def convertDateTimeToUnix2(inputDateTime):
#     outputDateTime = pd.to_datetime(inputDateTime).astype("int64")//1e9
#     return outputDateTime

# # Function to convert datetime (dtype: unix/epoch) back to dtype: datetime64
# def convertUnixDateTimeToDateTime64(inputDateTime):
#     outputDateTime = pd.to_datetime(inputDateTime, unit="s")
#     return outputDateTime
    
```

#### Create Start and Stop Stations DataFrames; then union/concat together

##### Start Stations

```python
# Start Stations
start_stations_df = cleanedCitiBike_df[["starttime", "start station id","start station name",\
                                       "start station latitude","start station longitude"]]

# Rename Start Stations Columns
start_stations_df = start_stations_df.rename(columns = {"starttime": "datetime",\
                                                        "start station id": "station id",\
                                                        "start station name": "station name",\
                                                        "start station latitude": "station latitude",\
                                                        "start station longitude": "station longitude"})

# Sort first by station name (asc) then datetime (desc) to be able to flag most recent event by station name
start_stations_df = start_stations_df.sort_values(["station name","datetime"],\
                                                  ascending=[True,False])

# Add column: "datetime_unix" to sort later on (by numeric value) and flag a row as most recent record per station
start_stations_df["datetime_unix"] = convertDateTimeToUnix(start_stations_df["datetime"])

# Add column: "startRankRow" to add row number per/partition by "station name", ordered by datetime (desc)
start_stations_df["startRankRow"] = start_stations_df.groupby("station name")["datetime_unix"]\
                                        .rank(method="first", ascending=False)\
                                        .astype(int)

# Add column: "startRankRowReverse" to add row number per/partition by "station name", ordered by datetime (asc)
start_stations_df["startRankRowReverse"] = start_stations_df.groupby("station name")["datetime_unix"]\
                                            .rank(method="first", ascending=True)\
                                            .astype(int)

# Add column: "startMostRecentFlag" to flag the most recent record for each "station name"
start_stations_df.loc[start_stations_df["startRankRow"] == 1, "startMostRecentFlag"] = 1
start_stations_df.loc[start_stations_df["startRankRow"] != 1, "startMostRecentFlag"] = 0
start_stations_df["startMostRecentFlag"] = start_stations_df["startMostRecentFlag"].astype(int)

# Reset Index to ascending order
start_stations_df = start_stations_df.reset_index()
del start_stations_df["index"]

# Vizualize the DataFrame
start_stations_df
```

#### Create DataFrame holding most recent station id, station latitude/longitude values along with the counts of Start and Stop records for each station

##### NOTE: there are some stations that do not have any start records - this is becuase a trip's journey began with a station in the area CVS data files were obtained through, and ended in an area that wasn't (i.e., started in NYC, ended in Jersey City)

```python
# Create start_stations_mrv_df on filtered values of start_stations_df for rows where startMostRecentFlag = 1
start_stations_mrv_df = start_stations_df.loc[start_stations_df['startMostRecentFlag'] == 1].reset_index()

# Drop unneeded columns
start_stations_mrv_df = start_stations_mrv_df.drop(columns=["index","datetime","datetime_unix","startRankRow","startMostRecentFlag"])

# Create stop_stations_mrv_df on filtered values of stop_stations_df for rows where stopMostRecentFlag = 1
stop_stations_mrv_df = stop_stations_df.loc[stop_stations_df['stopMostRecentFlag'] == 1].reset_index()

# Drop unneeded columns
stop_stations_mrv_df = stop_stations_mrv_df.drop(columns=["index","datetime","datetime_unix","stopRankRow","stopMostRecentFlag"])

# Merge/join start_stations_mrv_df and stop_stations_mrv_df
stations_mrv_df = pd.merge(start_stations_mrv_df, stop_stations_mrv_df, how="outer", on=["station name"])

# Replace NaN values in DataFrame with "NULL"
stations_mrv_df = stations_mrv_df.fillna("NULL")

# Use "station id" from start_stations_df, unless NaN / "NULL", then use stop_stations_df
stations_mrv_df.loc[stations_mrv_df["station id_x"] != "NULL", "station id"] = stations_mrv_df["station id_x"]
stations_mrv_df.loc[stations_mrv_df["station id_x"] == "NULL", "station id"] = stations_mrv_df["station id_y"]

# Use "station latitude" from start_stations_df, unless NaN / "NULL", then use stop_stations_df
stations_mrv_df.loc[stations_mrv_df["station latitude_x"] != "NULL", "station latitude"] = stations_mrv_df["station latitude_x"]
stations_mrv_df.loc[stations_mrv_df["station latitude_x"] == "NULL", "station latitude"] = stations_mrv_df["station latitude_y"]

# Use "station longitude" from start_stations_df, unless NaN / "NULL", then use stop_stations_df
stations_mrv_df.loc[stations_mrv_df["station longitude_x"] != "NULL", "station longitude"] = stations_mrv_df["station longitude_x"]
stations_mrv_df.loc[stations_mrv_df["station longitude_x"] == "NULL", "station longitude"] = stations_mrv_df["station longitude_y"]

# Use "startRankRowReverse" from start_stations_df, unless NaN / "NULL", then 0
stations_mrv_df.loc[stations_mrv_df["startRankRowReverse"] != "NULL", "stationStartRecords"] = stations_mrv_df["startRankRowReverse"]
stations_mrv_df.loc[stations_mrv_df["startRankRowReverse"] == "NULL", "stationStartRecords"] = 0

# Use "stopRankRowReverse" from start_stations_df, unless NaN / "NULL", then 0
stations_mrv_df.loc[stations_mrv_df["stopRankRowReverse"] != "NULL", "stationStopRecords"] = stations_mrv_df["stopRankRowReverse"]
stations_mrv_df.loc[stations_mrv_df["stopRankRowReverse"] == "NULL", "stationStopRecords"] = 0

# Drop unneeded columns
stations_mrv_df = stations_mrv_df.drop(columns=["station id_x","station latitude_x","station longitude_x","startRankRowReverse","station id_y","station latitude_y","station longitude_y","stopRankRowReverse"])

# Visualize the DataFrame
stations_mrv_df
```
